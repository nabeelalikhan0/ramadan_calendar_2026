```python
import json

INPUT_FILE = "timings.json"      # your saved API response
OUTPUT_FILE = "clean_ramadan.json"   # cleaned output


def clean_time(t):
    """Remove timezone text like ' (UTC)'"""
    return t.split(" ")[0]


with open(INPUT_FILE, "r", encoding="utf-8") as f:
    raw = json.load(f)

cleaned = []

for day in raw["data"]:
    timings = day["timings"]
    hijri = day["date"]["hijri"]
    greg = day["date"]["gregorian"]

    cleaned.append({
        "gregorian_date": greg["date"],
        "weekday": greg["weekday"]["en"],
        "hijri_date": hijri["date"],
        "ramadan_day": hijri["day"],
        "prayers": {
            "Imsak": clean_time(timings["Imsak"]),
            "Fajr": clean_time(timings["Fajr"]),
            "Sunrise": clean_time(timings["Sunrise"]),
            "Dhuhr": clean_time(timings["Dhuhr"]),
            "Asr": clean_time(timings["Asr"]),
            "Maghrib": clean_time(timings["Maghrib"]),
            "Isha": clean_time(timings["Isha"]),
        }
    })


with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(cleaned, f, indent=2, ensure_ascii=False)

print(f"✅ Clean data saved to {OUTPUT_FILE}")
```
